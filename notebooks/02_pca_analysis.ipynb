```
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
df=pd.read_csv('../data/heart.csv')
X=df.drop(columns=['target']).select_dtypes(include=[int,float]).fillna(df.median(numeric_only=True))
Z=StandardScaler().fit_transform(X)
pca=PCA().fit(Z)
import numpy as np
cum=np.cumsum(pca.explained_variance_ratio_)
plt.figure(); plt.plot(range(1,len(cum)+1), cum, marker='o'); plt.xlabel('n_components'); plt.ylabel('cumulative explained variance'); plt.grid(True); plt.tight_layout(); plt.savefig('../results/pca_cumulative.png', dpi=140)
```
