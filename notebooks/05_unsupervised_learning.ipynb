```
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage, dendrogram
df=pd.read_csv('../data/heart.csv')
y=df['target']
x=df.drop(columns=['target']).select_dtypes(include=[int,float]).fillna(df.median(numeric_only=True))
Z=StandardScaler().fit_transform(x)
km=KMeans(n_clusters=3,n_init=10,random_state=42).fit(Z)
labels=km.labels_
pca=PCA(n_components=2,random_state=42).fit_transform(Z)
import pandas as pd
pd.DataFrame({'pc1':pca[:,0],'pc2':pca[:,1],'cluster':labels,'target':y}).to_csv('../results/kmeans_summary.csv',index=False)
plt.figure(); plt.scatter(pca[:,0],pca[:,1],c=labels); plt.tight_layout(); plt.savefig('../results/pca_scatter.png',dpi=140)
L=linkage(Z,method='ward')
plt.figure(figsize=(10,4)); dendrogram(L,truncate_mode='level',p=5,no_labels=True); plt.tight_layout(); plt.savefig('../results/hierarchical_dendrogram.png',dpi=140)
```
